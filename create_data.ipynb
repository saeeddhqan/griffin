```python
def create_boundaries(continuous_vals, min_value, max_value, num_buckets):
    bucket_width = (max_value - min_value) / num_buckets
    boundaries = torch.empty(num_buckets + 1)
    boundaries[0] = min_value
    print(bucket_width)
    for i in range(1, num_buckets + 1):
        boundaries[i] = boundaries[i - 1] + bucket_width
    categories = torch.bucketize(continuous_vals, boundaries)
    return categories

def concat(a, b, c, d, e, f, g, h, i, j, k, l, m, n, o, p):
    concatenated_tensor = torch.cat((
        a.unsqueeze(1), b.unsqueeze(1), c.unsqueeze(1), d.unsqueeze(1),
        e.unsqueeze(1), f.unsqueeze(1), g.unsqueeze(1), h.unsqueeze(1),
        i.unsqueeze(1), j.unsqueeze(1), k.unsqueeze(1), l.unsqueeze(1),
        m.unsqueeze(1), n.unsqueeze(1), o.unsqueeze(1), p.unsqueeze(1)), dim=1)
    results = concatenated_tensor.view(-1)
    return results

def concat3(a, b, c, d, e, f, g, h):
    concatenated_tensor = torch.cat((
        a.unsqueeze(1), b.unsqueeze(1), c.unsqueeze(1), d.unsqueeze(1),
        e.unsqueeze(1), f.unsqueeze(1), g.unsqueeze(1), h.unsqueeze(1),
        i.unsqueeze(1), j.unsqueeze(1), k.unsqueeze(1), l.unsqueeze(1),
        m.unsqueeze(1), n.unsqueeze(1), o.unsqueeze(1), p.unsqueeze(1)), dim=1)
    results = concatenated_tensor.view(-1)
    return results


def concat2(a, b, c, d):
    concatenated_tensor = torch.cat((a.unsqueeze(1), b.unsqueeze(1), c.unsqueeze(1), d.unsqueeze(1)), dim=1)
    results = concatenated_tensor.view(-1)
    return results
```
